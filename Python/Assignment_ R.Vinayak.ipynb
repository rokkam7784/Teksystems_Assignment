```python
## Approach 2
def Withdraw():
    ## Part 1 Taking Denomination Input
    while True :
        try :
            print("Give the Count of the Number of notes for Each denominations:")
            denomList = ["note2000","note500","note100","note50","note20","note10","note5"]
            denoms = [2000,500,100,50,20,10,5]
            count = []
            for i in denomList:
                count.append(int(input("Num of notes of Rs."+i+" :")))


            denom_count_dir = {i:j for i,j in zip(denoms,count)}
            break
        except :
            print("! ERROR IN THE INPUT OCCURED !")
            print("! TRY AGAIN !")
            print("\n\n")
    
    while True :
            amount = int(input("Amount needs to be withdrawn :"))
            if amount>=0:
                break 
            
    
    ## Part 2 Function to find the combinations
    def combSum(denom_count_dir,total):
        comb,temp = [],[]
        denoms = sorted(list(set(denom_count_dir.keys())))
        findComb(comb,denoms,denom_count_dir,temp,total,0)
        return comb

    def findComb(comb,denoms,denom_count_dir,temp,total,index):
        if total==0:
            comb.append(list(temp))
            return
        for i in range(index,len(denoms)):
            org_dict =  denom_count_dir.copy()
            if denom_count_dir[denoms[i]] >= 0 :
                if total - denoms[i] >=0:
                    denom_count_dir[denoms[i]] -= 1
                    temp.append(denoms[i])
                    findComb(comb,denoms,denom_count_dir,temp,total-denoms[i],i)
                    temp.remove(denoms[i])
            denom_count_dir = org_dict
    
    
    org_dict = denom_count_dir.copy()
                
    countL = []
    for i in range(len(denoms)):
        countL.extend([denoms[i] for j in range(count[i])])
    print("There are total",len(countL),"notes in total")
    print()
    
    # Part 3 Executing and finding combinations
    comb = combSum(denom_count_dir,amount)
    
    
    if len(comb)==0:
        print("Empty !")
#     print(comb)
    for i in comb:
        show = True
        denom_count_dir = org_dict
        for j in set(i):
#             print( i.count(j) , denom_count_dir[j])
            if i.count(j) > denom_count_dir[j]:
#                 print( False)
                show = False
                
        if show :
            for j in set(i):
                print("Requires",i.count(j),j,"Note")
            print()
#     print(denom_count_dir)
```
