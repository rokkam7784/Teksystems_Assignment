```python
def sumOfDigits(s):
    s = str(s)
    sm = 0
    for e in s:
        sm += int(e)
    return sm
```

```python
def countDividedBy():
    n = int(input("N : "))
    
    l = []
    for _ in range(n):
        l.append(int(input()))
    
    count = 0
    for i in l:
        if i%3==0 and i%4==0 and sumOfDigits(i)%5!=0:
            count+=1
    
    return count
    
countDividedBy()  
```

```python
## ans: 
def isAnagrams(s1,s2):
    s1,s2 = list(s1.lower()),list(s2.lower())
    for i in s1:
        if s2.count(i)!=s1.count(i):
            return False
    return True
```

```python
isAnagrams("care","raace")
```

```python
isAnagrams("care","race")
```

```python
def StringBinaryOperations(s1):
    for i in s1:
        if i == "Z":
            s1 = s1.replace("Z","^")
        elif i == "Y":
            s1 = s1.replace("Y","|")
        elif i == "X":
            s1 = s1.replace("X","&")
    print(s1,"=",eval(s1))
    
    
StringBinaryOperations("1Z0Z1Z1X0Y1")
```

```python
def findAutoCount(n):
    if n == None :
        return 0
    
    desired = True

    list_n = list(n)
    for i in range(len(list_n)):
        if int(list_n[i])!= list_n.count(str(i)):
#             print(list_n[i],list_n.count(str(i)))
            desired = False
            
    if desired:
        return len(set(n))
    else:
        return 0
    
findAutoCount("1210")
```

```python
findAutoCount("21200")
```

```python
findAutoCount(None)
```
